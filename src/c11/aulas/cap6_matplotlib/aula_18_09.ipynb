```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # Graficos fundamentais


x = np.array([1,2,3,4])
y = x * 2
y2 = x * x

plt.xlabel('Valores de X')
plt.ylabel('Valores de y')
# Como os dados alteram ao longo do tempo
# Criando a matriz de subplot, dashboard
plt.subplot(1, 2, 1)
plt.plot(x, y, '*:r', linewidth=3, markersize=10)

plt.subplot(1, 2, 2)
plt.plot(x, y2, 's--g', linewidth=3, markersize=10)
# os dois graficos juntos
# plt.plot(x, y, 'h:r',x, y2,'s--g', linewidth=3, markersize=10)
plt.show()
```

```python
df = pd.read_csv('/home/nkk/rea de trabalho/projetos/C11/src/c11/aulas/paises.csv', sep=';')
df.head()
# Scatter(disperso) 
# Pegando os seis maiores paises em area
# n maiores paises em area
maiores_paises_area = df.nlargest(6, 'Area (sq. mi.)')
print(maiores_paises_area['Country'])
# print(maiores_paises_area['GDP ($ per capita)'])
# qual longe estao desses pontos
# Tracar um scatter plot que ilustre o GDP per capita destes paises
plt.scatter(maiores_paises_area['Country'], maiores_paises_area['GDP ($ per capita)'], s=300)
plt.show()
```

```python
cinco_maiores_paises_pib = df.nlargest(5, 'GDP ($ per capita)')
# print(cinco_maiores_paises_pib['Country'])
# Comparar valor
plt.bar(cinco_maiores_paises_pib['Country'], cinco_maiores_paises_pib['GDP ($ per capita)'])
plt.show()
```

```python
paises_sem_costa = df[df['Coastline (coast/area ratio)'] == 0]

# Extraindo quando paises nao tem costa
qtd_sem_costa = len(paises_sem_costa)
qtd_com_costa = len(df) - qtd_sem_costa

# plotando o grafico de pizza
# autopct mostrar em porcetagem - proporo
plt.pie([qtd_sem_costa, qtd_com_costa], labels=['Paises sem costa', 'Paises com costa'], autopct='%1.1f%%')
plt.show()
```
