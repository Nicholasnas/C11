```python
"""""
    Por meio do dataset paises.csv, trace dois gráficos de linhas em um mesmo plano
    cartesiano, um mostrando a taxa de mortalidade (Deathrate) e outro a taxa de natalidade (Birthrate) dos países da
    América do Norte """


import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("/home/nkk/Área de trabalho/projetos/C11/src/c11/aulas/paises.csv", sep=';')

paises_america_norte = df[df['Region'].str.contains('NORTHERN AMERICA')]

# Plotar gráficos de linhas no mesmo plano
plt.figure(figsize=(10,6))

plt.plot(paises_america_norte['Country'], paises_america_norte['Deathrate'], 
         marker='o', label='Taxa de Mortalidade')

plt.plot(paises_america_norte['Country'], paises_america_norte['Birthrate'], 
         marker='o', label='Taxa de Natalidade')

plt.xlabel("Países")
plt.ylabel("Taxa")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Exercicio 2

space = pd.read_csv("/home/nkk/Área de trabalho/projetos/C11/space.csv", sep=";")

empresas_usa = space[space["Location"].str.contains("USA")]["Company Name"].nunique()
empresas_china = space[space["Location"].str.contains("China")]["Company Name"].nunique()

df_empresas = pd.DataFrame({
    "País": ["EUA", "China"],
    "Qtd Empresas": [empresas_usa, empresas_china]
})

plt.figure(figsize=(6,5))
plt.bar(df_empresas["País"], df_empresas["Qtd Empresas"], color=["blue", "red"])
plt.title("EUA vs China")
plt.ylabel("Quantidade de Empresas")
plt.show()
```

```python
# Exercicio 3

# Agencia espacial roscosmos
roscosmos = space[space["Company Name"] == "Roscosmos"]

# Conta status das missoes
status = roscosmos["Status Mission"].value_counts()

# Pizza
plt.figure(figsize=(6,6))
plt.pie(status, labels=status.index, autopct="%.1f%%", startangle=90, colors=["green","red"])
plt.title("Taxa de sucesso vs fracasso - Roscosmos")
plt.show()
```
