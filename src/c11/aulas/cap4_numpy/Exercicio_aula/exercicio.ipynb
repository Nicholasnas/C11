```python
# 4 - Usando np.unique, conte quantos países existem por região (semelhante ao seu exercício, 
# mas agora mostrando também a porcentagem em relação ao total de países).
regiao, quantidade = np.unique(regioes, return_counts=True)

total_de_paises = np.sum(quantidade)
print(f'Total de paises no mundo: {total_de_paises}')

for region, count in zip(regiao, quantidade):
    print(f"A regiao '{region}' tem {count} países.")
    print(f'Porcentagem em relação ao total de países: {(count/total_de_paises)*100:.2f}%', end='\n\n')
```

```python
# Coluna de Regiões (Region)
regioes = df[1:, 1]  

# Coluna de alfabetização (Literacy %) convertida para float
# Obs: literacy aqui ainda é de TODO o dataset, vamos refinar no subconjunto depois
literacy = df[1:, 9].astype(float)  

# --- Passo 1: filtrar só países cuja região contém "EUROPE"
# np.char.find(regioes, "EUROPE") retorna o índice da substring se encontrada, ou -1 se não.
# Então com '!= -1' transformamos em uma máscara booleana (True para quem contém "EUROPE").
# Aplicando essa máscara em df[1:, :] pegamos TODAS as colunas, mas só das linhas da Europa.
paises_europe = df[1:, :][np.char.find(regioes, "EUROPE") != -1]

# --- Passo 2: agora trabalhamos só com esse subconjunto
# Extraímos a coluna de alfabetização (coluna 9) dentro do subconjunto europeu
literacy_europe = paises_europe[:, 9].astype(float)

# --- Passo 3: aplicar a condição
# Criamos uma máscara booleana dentro da Europa: True apenas para países com Literacy > 95
# Usamos essa máscara para selecionar apenas a coluna 0 (Country) do subconjunto
country_europe_good = paises_europe[literacy_europe > 95, 0]

# --- Passo 4: saída
print("Países da Europa com alfabetização > 95%:")
print(country_europe_good)
```
