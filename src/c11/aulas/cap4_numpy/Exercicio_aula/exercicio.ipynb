```python
# 4 - Usando np.unique, conte quantos pa√≠ses existem por regi√£o (semelhante ao seu exerc√≠cio, 
# mas agora mostrando tamb√©m a porcentagem em rela√ß√£o ao total de pa√≠ses).
regiao, quantidade = np.unique(regioes, return_counts=True)

total_de_paises = np.sum(quantidade)
print(f'Total de paises no mundo: {total_de_paises}')

for region, count in zip(regiao, quantidade):
    print(f"A regiao '{region}' tem {count} pa√≠ses.")
    print(f'Porcentagem em rela√ß√£o ao total de pa√≠ses: {(count/total_de_paises)*100:.2f}%', end='\n\n')
```

```python
# Coluna de Regi√µes (Region)
regioes = df[1:, 1]  

# Coluna de alfabetiza√ß√£o (Literacy %) convertida para float
# Obs: literacy aqui ainda √© de TODO o dataset, vamos refinar no subconjunto depois
literacy = df[1:, 9].astype(float)  

# --- Passo 1: filtrar s√≥ pa√≠ses cuja regi√£o cont√©m "EUROPE"
# np.char.find(regioes, "EUROPE") retorna o √≠ndice da substring se encontrada, ou -1 se n√£o.
# Ent√£o com '!= -1' transformamos em uma m√°scara booleana (True para quem cont√©m "EUROPE").
# Aplicando essa m√°scara em df[1:, :] pegamos TODAS as colunas, mas s√≥ das linhas da Europa.
paises_europe = df[1:, :][np.char.find(regioes, "EUROPE") != -1]

# --- Passo 2: agora trabalhamos s√≥ com esse subconjunto
# Extra√≠mos a coluna de alfabetiza√ß√£o (coluna 9) dentro do subconjunto europeu
literacy_europe = paises_europe[:, 9].astype(float)

# --- Passo 3: aplicar a condi√ß√£o
# Criamos uma m√°scara booleana dentro da Europa: True apenas para pa√≠ses com Literacy > 95
# Usamos essa m√°scara para selecionar apenas a coluna 0 (Country) do subconjunto
country_europe_good = paises_europe[literacy_europe > 95, 0]

# --- Passo 4: sa√≠da
print("Pa√≠ses da Europa com alfabetiza√ß√£o > 95%:")
print(country_europe_good)
```
